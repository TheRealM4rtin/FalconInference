```python
query = "J'ai besoin d'une équipe de 3 à 4 intervenants pour une mission banque pour le développement d'une digital factory"
```

```python
def prompt(query):
  prompt = """ '''
      ----------------------------------------
      Confidential work here
      ----------------------------------------

        Format attendu :

        n : -



        Quand tu ne trouves pas l'information, tu dois indiquer la mention "Non renseigné"




        Exemple de description d'équipe :

        2 à 3 intervenants




        Exemple de réponse attendue :

        n : 3

'''

        Description d'équipe:

        """ + query
  return prompt
```

```python
query = query
```

```python
prompt = prompt(query)
```

```python
print(prompt)
```

```python
for pr in list_query_from_chatgpt:
  sequences = pipeline(prompt(pr),
    max_length=300,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
  )

for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```
